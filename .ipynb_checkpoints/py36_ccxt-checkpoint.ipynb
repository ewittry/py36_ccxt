```python
import ccxt
import json
from pprint import pprint
import requests
from requests.auth import AuthBase
from ddlgenerator.ddlgenerator import Table
from sqlalchemy.sql import text
from sqlalchemy import create_engine
import pandas as pd
import pandas.io.sql as psql
import hmac
import time
import hashlib
import Crypto
import base64
#from bittrex import bittrex
```

```python
url = "https://bittrex.com/api/v1.1/public/getmarketsummaries"
r = requests.post(url)  
#print (r)
data = json.loads(r.content.decode())
#print (data)
```

```python
table = Table(data['result'])
sql = table.sql('postgresql', inserts=True)
```

```python
#sql
```

```python
sql = sql.split('\n\n')
sql[2] = sql[2].split('\n')
mydb_lst = sql[0].split(' ')
mydb = mydb_lst[4]
mydb = mydb[:-1]
```

```python
# create the session
engine = create_engine('sqlite:///:memory:')
```

```python
res = engine.execute(sql[0])
res = engine.execute(sql[1])
for x in sql[2]:
    res = engine.execute(x)
```

```python
t = text("SELECT * FROM " + mydb + ";")
t = text("SELECT * FROM " + mydb + " WHERE marketname like 'BTC%' order by _timestamp asc;")
res = engine.execute(t)
```
